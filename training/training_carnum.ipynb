```python
cap = cv2.VideoCapture('carnum_test.mp4')

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

#   XVID
fourcc = cv2.VideoWriter_fourcc(*'XVID')
# 30 XVID    
writer = cv2.VideoWriter('carnum_0921_1_final.avi', fourcc, 30.0, (width, height))
while(cap.isOpened()) :
    ret, frame = cap.read()
    if ret == False:
        break
    result = tfnet.return_predict(frame)
    result_img = frame.copy()
    #   
    if result:
        print(result)
        for res in result:
            x = res['topleft']['x']
            y = res['topleft']['y']
            w = res['bottomright']['x']
            h = res['bottomright']['y']
            target = frame[y:h, x:w]
            flt = cv2.blur(target, (60, 60))  # blur   flt  
            result_img[y:h, x:w] = flt
            tl = (x, y)
            br = (w ,h)
            label = res['label']
            conf = res['confidence']
            #result_img= cv2.rectangle(result_img, tl, br, (0, 255, 0), 2)
            #result_img= cv2.putText(result_img, label, tl, cv2.FONT_HERSHEY_TRIPLEX, 1, (0,0,0), 2)
        
    else :
        result_img = frame.copy()
        

    writer.write(result_img)

    if cv2.waitKey(1)&0xFF == 27:
        break
cv2.destroyAllWindows()
```
